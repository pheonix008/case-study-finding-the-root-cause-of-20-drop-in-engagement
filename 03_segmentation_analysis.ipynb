```python
import pandas as pd

df = pd.read_csv("../data/criteo_ctr_drop_dataset.csv")

df['date'] = pd.to_datetime(df['date'])
df['ctr'] = df['clicks']/df['impressions']
```

```python
before = df[df['date'] < '2026-06-15']
```

```python
after = df[df['date'] >= '2026-06-15']
```

```python
device_before = (
    before.groupby('device')
    .agg({'clicks':'sum',
          'impressions':'sum'})
)

device_before['ctr_before'] = (
    device_before['clicks']
    /
    device_before['impressions']
)

device_before
```

```python
device_after = (
    after.groupby('device')
    .agg({'clicks':'sum',
          'impressions':'sum'})
)

device_after['ctr_after'] = (
    device_after['clicks']
    /
    device_after['impressions']
)

device_after
```

```python
device_analysis = pd.concat(
    [
        device_before['ctr_before'],
        device_after['ctr_after']
    ],
    axis=1
)

device_analysis
```

```python
mobile_before = before[
    before['device']=='Mobile'
]

mobile_after = after[
    after['device']=='Mobile'
]
mobile_after
```

```python
browser_before=( mobile_before.groupby('browser').agg({'clicks':'sum','impressions':'sum'}))
```

```python
browser_before
```

```python
browser_before['ctr_before']=browser_before['clicks']/browser_before['impressions']
```

```python
browser_before
```

```python
browser_after=(mobile_after.groupby('browser').agg({'clicks':'sum','impressions':'sum'}))
browser_after['ctr_after']=browser_after['clicks']/browser_before['impressions']
```

```python
browser_after
```

```python
browser_analysis = pd.concat(
    [
        browser_before['ctr_before'],
        browser_after['ctr_after']
    ],
    axis=1
)

browser_analysis
```
