```python
control_clicks = 9000
control_impressions = 300000

treatment_clicks = 15000
treatment_impressions = 300000
```

```python
control_ctr = control_clicks/control_impressions

treatment_ctr = (
    treatment_clicks
    /
    treatment_impressions
)

print(control_ctr)
print(treatment_ctr)
```

```python
from statsmodels.stats.proportion import proportions_ztest

clicks = [
    control_clicks,
    treatment_clicks
]

impressions = [
    control_impressions,
    treatment_impressions
]

z_stat,p_value = proportions_ztest(
    clicks,
    impressions
)

print(z_stat)
print(p_value)
```
