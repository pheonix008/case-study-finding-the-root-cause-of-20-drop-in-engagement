```python
import pandas as pd

df = pd.read_csv("../data/criteo_ctr_drop_dataset.csv")

df['date'] = pd.to_datetime(df['date'])
df['ctr'] = df['clicks']/df['impressions']
```

```python
mobile_safari = df[
    (df['device']=='Mobile')
    &
    (df['browser']=='Safari')
]
```

```python
before_ms = mobile_safari[
    mobile_safari['date']<'2026-06-15'
]

after_ms = mobile_safari[
    mobile_safari['date']>='2026-06-15'
]
```

```python
before_clicks = before_ms['clicks'].sum()
```

```python
after_clicks = after_ms['clicks'].sum()
```

```python
print(before_clicks)
print(after_clicks)
```

```python
before_impressions = before_ms['impressions'].sum()
after_impressions = after_ms['impressions'].sum()

print(before_impressions)
print(after_impressions)
```

```python
before_ctr = before_clicks / before_impressions
after_ctr = after_clicks / after_impressions

print(before_ctr)
print(after_ctr)
```

null hypothesis:CTR_before = CTR_after

alternate hypothesis:CTR_before != CTR_after

```python
from statsmodels.stats.proportion import proportions_ztest

clicks = [
    before_clicks,
    after_clicks
]

impressions = [
    before_impressions,
    after_impressions
]

z_stat, p_value = proportions_ztest(
    clicks,
    impressions
)

print("Z-statistic =", z_stat)
print("P-value =", p_value)
```
